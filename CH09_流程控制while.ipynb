# 基本使用

```python
sum = 0
while sum <= 100:
    sum += 1

print(sum)
```

```python
name_list = list()

name = ''
while name != 'q':
    name = input("輸入要儲存的名字，結束請按q")
    if name != 'q':
        name_list.append(name)

print("你所輸入的名字有: ",name_list)
```

```python
price = 50
inflation_rate = 0.03
year = 0

while price < 75:
    price = price * (1 + inflation_rate)
    year += 1
    
print("過了",year,'年')
print("商品價格將上漲為",price)
```

```python
sum = 0
while 1:#無限迴圈
    sum += 1
print(sum)
```

# 搭配Break

```python
price = 50
inflation_rate = 0.03
year = 0

while True:
    price = price * (1 + inflation_rate)
    year += 1
    
    if price > 75:
        break
    
print("過了",year,'年')
print("商品價格將上漲為",price)
```

# 使用旗標

```python
price = 50
inflation_rate = 0.03
year = 0
flag = True

while flag is True:
    price = price * (1 + inflation_rate)
    year += 1
    
    if price > 75:
        flag = False
    
print("過了",year,'年')
print("商品價格將上漲為",price)
```

# continue的使用

```python
price = 50
inflation_rate = 0.03
year = 0
flag = True

while flag is True:
    price = price * (1 + inflation_rate)
    year += 1
    
    if price <= 75:
        continue
        
    flag = False
    print("過了",year,'年')
    print("商品價格將上漲為",price)
```

```python
a = 35

while a > 0:
    a -= 2
    continue
    print("123")
    
print("456")
    
```

# pass的使用

```python
price = 50
inflation_rate = 0.03
year = 0
flag = True

while flag is True:
    price = price * (1 + inflation_rate)
    year += 1
    
    if price <= 75:
        pass
        print("after pass")
        
    flag = False
    print("過了",year,'年')
    print("商品價格將上漲為",price)
```

# pass可以放在未完成的流程區塊

```python
a = 35

if a > 0:
    pass
elif a > 100:
    pass
elif a > 1000:
    pass
else:
    pass
```
