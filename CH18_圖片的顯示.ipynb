# 更改圖像的尺寸(放大或縮小)

### 方法一:使用設定數值

```python
img_bgr = cv2.imread(img_path)

if img_bgr is None:
    print("Read failed:",img_path)
else:
    #----BGR to RGB
    img_rgb = img_bgr[:,:,::-1]
    
    #----resize 
    height = 100
    width = 120
    
    img_resized = cv2.resize(img_rgb,(width,height))
    
    print("image shape: ",img_resized.shape)
    
    #----display the image
    plt.imshow(img_resized)
    plt.show()
```

### 方法二:使用設定比例

```python
img_bgr = cv2.imread(img_path)

if img_bgr is None:
    print("Read failed:",img_path)
else:
    #----BGR to RGB
    img_rgb = img_bgr[:,:,::-1]
    
    #----resize
    img_resized = cv2.resize(img_rgb,None,fx=0.8,fy=0.7)
    
    print("image shape: ",img_resized.shape)
    
    #----display the image
    plt.imshow(img_resized)
    plt.show()
```

# 練習

```python
import random
```

```python
dir_path = r"F:\book_examples\lfw"

#----var
path_list = list()
crop_height = 175
crop_width = 175

#----get all paths
for dir_name,sub_dir_list,filename_list in os.walk(dir_path):
    if len(filename_list) > 0:
        for filename in filename_list:
            if filename.split(".")[-1] == 'jpg':
                path = os.path.join(dir_name,filename)
                path_list.append(path)
                        
print("總共有幾張圖片:",len(path_list))

#----random selection
random.shuffle(path_list)

#----create the output dir
save_dir = os.path.join(os.path.dirname(dir_path),'random_selection')
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
    
#----read images
for path in path_list[:50]:
    img = cv2.imread(path)
    if img is None:
        print("Read failed:",path)
    else:
        #----crop
        height,width,_ = img.shape
        center_height = int(height / 2)
        h_start = center_height - int(crop_height / 2) 
        h_end = h_start + crop_height
        #print("h_start = {}, h_end = {}".format(h_start,h_end))

        center_width = int(width / 2)
        w_start = center_width - int(crop_width / 2) 
        w_end = w_start + crop_width
        #print("w_start = {}, w_end = {}".format(w_start,w_end))
        img_crop = img[h_start:h_end, w_start:w_end,:]
        
        #----save images
        splits = path.split("\\")[-1].split(".")
        filename = "{}_crop.{}".format(splits[0],splits[-1])
        new_path = os.path.join(save_dir,filename)
        
        cv2.imwrite(new_path,img_crop)
        print("{}儲存成功".format(new_path))
        
        
        
```

# 建立四維圖片資料

## 方式1:使用串列容器接收圖片資料

```python
dir_path = r"D:\dataset\lfw_1"

#----var
path_list = list()
img_data = list()

#----get all paths
for dir_name,sub_dir_list,filename_list in os.walk(dir_path):
    if len(filename_list) > 0:
        for filename in filename_list:
            if filename.split(".")[-1] == 'jpg':
                path = os.path.join(dir_name,filename)
                path_list.append(path)
                        
print("總共有幾張圖片:",len(path_list))

#----read images to form 4D data
for path in path_list[:50]:
    img = cv2.imread(path)
    if img is None:
        print("Read failed:",path)
    else:
        img_data.append(img)#資料型態是串列(list)
        
#----轉換成numpy array
img_data = np.array(img_data)

#----check 4D data info
print(img_data.shape)
print(img_data.dtype)
```

## 方式2:使用ndarray接收圖片資料

```python
dir_path = r"D:\dataset\lfw_1"

#----var
path_list = list()
read_qty = 50

#----宣告numpy array的容器
img_data = np.zeros((read_qty,250,250,3),dtype=np.uint8)

#----get all paths
for dir_name,sub_dir_list,filename_list in os.walk(dir_path):
    if len(filename_list) > 0:
        for filename in filename_list:
            if filename.split(".")[-1] == 'jpg':
                path = os.path.join(dir_name,filename)
                path_list.append(path)
                        
print("總共有幾張圖片:",len(path_list))

#----read images to form 4D data
for i,path in enumerate(path_list[:read_qty]):
    img = cv2.imread(path)
    if img is None:
        print("Read failed:",path)
    else:
        img_data[i] = img

#----check 4D data info
print(img_data.shape)
print(img_data.dtype)
```

```python
def get_paths(dir_path):
    for dir_name,sub_dir_list,filename_list in os.walk(dir_path):
        if len(filename_list) > 0:
            for filename in filename_list:
                if filename.split(".")[-1] == 'jpg':
                    path = os.path.join(dir_name,filename)
                    path_list.append(path)
                        
    print("總共有幾張圖片:",len(path_list))
    return path_list
```

```python
dir_path = r"D:\dataset\lfw_1"

#----var
path_list = list()
img_data = list()
row_number = 2
column_number = 3

#----get all paths
path_list = get_paths(dir_path)

#----random selection
random.shuffle(path_list)

#----read images to form 4D data
for i,path in enumerate(path_list[:int(row_number*column_number)]):
    img = cv2.imread(path)
    if img is None:
        print("Read failed:",path)
    else:
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        img_data.append(img)

#----display
plt.subplot(3,2,1)
plt.imshow(img_data[0])

plt.subplot(3,2,2)
plt.imshow(img_data[1])

plt.subplot(3,2,3)
plt.imshow(img_data[2])

plt.subplot(3,2,4)
plt.imshow(img_data[3])

plt.subplot(3,2,5)
plt.imshow(img_data[4])

plt.subplot(3,2,6)
plt.imshow(img_data[5])

plt.show()
```
